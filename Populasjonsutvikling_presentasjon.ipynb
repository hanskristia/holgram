# Populasjonsutvikling

## Eksponentiel vekst

En art øker med 20% hvert år. Hvor mange individer er det av den arten om 10 år dersom startpopulasjonen er 10?

For å løse komplekse problemer bruker vi ofte en iterativ metode. Det vil si at vi ser på hvor mange individer det er nå, så regner vi ut hvor mange individer det blir neste år. 

Gode oppfølgningsoppgaver til elevene kan være, hvor lang tid tar det før populasjonen er over 10000? Da kan de eksempelvis endre while-løkken til while populasjon < 10000:. 

## Logistisk vekst
Hvis man lar forrige simulasjon gå i noen år ser man at antallet individer blir enorm. Dette er ikke helt realistisk. 

Vi ønsker at endringen av populasjonen skal gå mot 0 når populasjonen når en viss bæreevne. Dette kan vi oppnå med å la endringen hvert år være: 

$r\cdot populasjon(1-\frac{B}{populasjon})$,
hvor r er veksten og B er bæreevnen. 

Hvorfor funker dette? Og hvordan vil utviklingenvære dersom bæreevnen er 2000, r er 20% og startpopulasjonen er 10?

## To arter med samme nisje 
Hvis vi har to forskjellige arter som begge har samme nisje vil de selvfølgelig påvirke hverandre. Dette kan være at de spiser det samme eller at de tar opp det samme arealet. 

```python
pop_a = 100
pop_b = 50

k_a = 1000
k_b = 1000

r_a = 0.05
r_b = 0.05

ab = 0.2 #Hvor mye a påvirker b
ba = 0.3 #Hvor mye b påvirker a

tid = 0

a = [pop_a]
b = [pop_b]
t = [tid]

while tid < 1000: 
    pop_a = pop_a + r_a * pop_a*(1-(pop_a+pop_b*ba)/k_a)
    pop_b = pop_b + r_b * pop_b*(1-(pop_a*ab+pop_b)/k_b)
    
    
    a.append(pop_a)
    b.append(pop_b)
    tid +=1
    t.append(tid)
    if pop_a < 0 or pop_b < 0: 
        break
    
print(a[-1])
print(b[-1])
plt.figure()
plt.plot(t,a,label='pop_a')
plt.plot(t,b,label='pop_b')
plt.legend()
```

# Bytte og rovdyr (ekstra utfordring)

La oss ha en populasjon av kaniner på 1000 og rever på 50. For hver generasjon blir det 10% flere kaniner. Revene trenger mat for å forplante seg og spiser derfor kaniner. 



Antall muligheter for et byttedyr møter et rovdyr er antallet byttedyr multiplisert med antallet rovdyr. 

```python
tid_slutt = 100

pop_a = 100
pop_b = 10

r_a = 0.4
angrepsrate = 0.01
omgjøringsrate = 0.4
dødsrate = 0.3

tid = 0

a = [pop_a]
b = [pop_b]
t = [tid]

while tid < tid_slutt: 
    pop_a = pop_a + r_a*pop_a - pop_a*pop_b*angrepsrate 
    pop_b = pop_b + angrepsrate*omgjøringsrate*pop_a*pop_b - dødsrate*pop_b
    
    tid += 1
    
    a.append(pop_a)
    b.append(pop_b)
    t.append(tid)
    if pop_a < 0 or pop_b < 0:
        break
        
plt.figure()
plt.plot(t,a,label='bytte')
plt.plot(t,b,label='rovdyr')
plt.legend()
```
